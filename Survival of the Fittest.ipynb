#0. Intro
--------------------------------

The ongoing ["Bubble Talk"](http://blog.samaltman.com/bubble-talk) in today's startup ecosystem suffers from a profound lack of information surrounding startup death. A 2014 analysis by [CB Insights](https://www.cbinsights.com/blog/startup-death-data/) confirms that startup death is extremely hard to identify and plagued by survivorship bias. Unfortunately, without a clear understanding of how startups fail, it is nearly impossible to assess the risk of today's "Unicorns" turning into tomorrow's "Unicorpses" (terminology credit to [Aileen Lee](http://techcrunch.com/2015/07/18/welcome-to-the-unicorn-club-2015-learning-from-billion-dollar-companies) and [Brad Feld](http://www.feld.com/archives/2015/09/unicorpse.html)).

###Survival Analysis

With the understanding that imperfect knowledge is usually better than no knowledge at all, I decided to go ahead and see if I could dig up any insights on startup death using publicly available information from [`Crunchbase`](crunchbase.com) and survivor analysis. From Wikipedia:
>Survival analysis is a branch of statistics that deals with analysis of time duration until one or more events happen, such as death in biological organisms and failure in mechanical systems...Survival analysis attempts to answer questions such as: what is the proportion of a population which will survive past a certain time? Of those that survive, at what rate will they die or fail? Can multiple causes of death or failure be taken into account? How do particular circumstances or characteristics increase or decrease the probability of survival?

In other words, survivor analysis can help us predict and describe startup death using existing data on startup lifespans, failure rates, and relevant covariates. 

So, how is survivor analysis different from normal, run-of-the-mill linear regression? Well, unlike linear regression, SA can account for the missing data associated with subjects that live past our window of observation, a problem known as "left-sided censorship." Whereas standard regression needs to know the "Time of Death" (ToD) for every subject in the dataset to calculate failure rates, survival analysis interpolates the ToD for subjects that are still alive at the end of the observation window. Turning to the `Crunchbase` data, this means that we can include startups that are still operating in our analysis (rather than waiting for them to fail in order to find the ToD).

__tl;dr:__ Survivor analysis allows us to form credible insights using incomplete data.

The two key functions in survival analysis are the *survival function* and the *hazard function*.

The **survival function**, conventionally denoted by $S$, is the probability that the event (say, death) has not occurred yet:

$$S(t)=Pr(T>t)$$

where $T$ denotes the time of death and $Pr$ the probability. Since $S$ is a probability, $0≤S(t)≤1$. Survival times are non-negative $(T≥0)$ and, generally, $S(0)=1$.

The **hazard function** $h(t)$ is the event (death) rate at time $t$, conditional on survival until $t$ (i.e., $T≥t$):

$$h(t)= \lim_{\Delta t \rightarrow 0}\frac{Pr(t\leq T \leq t+\Delta t)}{S(t)}=\frac{p(t)}{S(t)}$$

where $p$ denotes the probability density function.

###The Kaplan–Meier Estimate

In practice, we do not get to observe the actual survival function of a population; we must use the observed data to estimate it. A popular estimate for the survival function $S(t)$ is the Kaplan–Meier estimate, which gives us a maximum-likelihood estimate of the survival function:
$$\hat{S}(t)=\prod_{t_i<t}\frac{n_i - d_i}{n_i}$$

where $d$ and $n$ are respectively the count of "death" events and individuals at risk at timestep $i$.

The cumulative product gives us a non-increasing curve where, at any timestep during the observation window, we can read off the estimated probability of survival from the start to that timestep. We can also compute the estimated survival time or median survival time (half-life).

###The Cox Proportional Hazard Estimator

The Cox PH model gives a semi-parametric method of estimating the hazard function $\lambda$ at time $t$ given a baseline hazard that's modified by a set of covariates:

$$\lambda(t\mid X)=\lambda_0(exp(\beta_1X_1+...+\beta_pX_p)= \lambda_0(exp(\beta X)$$

where $\lambda_0(t)$ is the non-parametric baseline hazard function and $\beta X$ is a linear parametric model using features of the individuals, transformed by an exponential function.

The algorithm runs quite quickly, performing iterative partial regression of the time-dependent $\lambda_0(t)$ and the feature-dependent $exp(\beta X)$, repeated until convergence.

The big strengths of Cox PH is that it allows us to:

+ compare the relative hazard rates of the different features by comparing the $\beta$ coefficients

+ compute a predicted survival function for new datapoints simply by plugging the details into the model. 
*(**Note:** the library in `Python` that handles survival analysis is buggy when it comes to Cox PH predictions. I've contacted the maintainer about updating some of the relevant functions, but we haven't resolved this issue yet. Hopefully I'll be able to update this analysis with predictions in the next couple weeks.)*

One big weakness of the Cox PH is that the $\beta$ values are assumed constant over all time. For continuous variables such as startup funding, this implies that the first dollar of VC capital has the same impact on failure rates as the millionth dollar. For categorical variables such as the market that a startup addresses, this implies that industry trends are static over time (i.e. markets don't grow, interact, or evolve). Clearly, neither of these assumptions are true; caveats aside, Cox PH still provides us with an extremely valuable way of prying apart causal effects among covariates.

###Local Functions

```python
#Remember all those workarounds I mentioned? These are some of them.
#Local is the new organic.
def estimate_cond_mean(S):
    """ Quick & dirty estimate of conditional mean lifetime """
    fstar = -S.diff() / (1-S.iloc[-1,0])
    Sstar = (S-S.iloc[-1,0]) / (1-S.iloc[-1,0])
    llstar = fstar / Sstar
  
    llstar[pd.isnull(llstar)] = 0
    llstar = llstar[np.isfinite(llstar)]
    llstarcs = llstar.cumsum().reset_index()
    llstarcs['timelinediff'] = np.append(llstarcs['timeline'].diff().values[1:],0)
    llstarcs['auc'] = llstarcs['timelinediff'] * llstarcs['KM_estimate']
    return np.nansum(llstarcs['auc']).round()

def plot_km(km, axes, suptxt='', subtxt='', i=0, j=0, arws=[], xmax=0, smlfs=10):

    ax = km.plot(ax=axes[i,j], title=subtxt, legend=False)
    plt.suptitle(suptxt, fontsize=14)
    axes[i,j].axhline(0.5, ls='--', lw=0.5)
    axes[i,j].annotate('half-life', fontsize=smlfs, color='b'
            ,xy=(0,0.5), xycoords='axes fraction'
            ,xytext=(10,4), textcoords='offset points')

    S = km.survival_function_
    hl = S.loc[S['KM_estimate']<0.5,'KM_estimate'].head(1)
    if len(hl) == 1:
        axes[i,j].annotate('{:.0f}'.format(hl.index[0]), fontsize=smlfs
            ,xy=(0,0.5), xycoords='axes fraction'
            ,xytext=(10,-12), textcoords='offset points', color='b')

    for pt in arws:
        tml = km.survival_function_[:pt].tail(1)
        plt.annotate('{:.1%}\n@ {:.0f}days'.format(tml['KM_estimate'].values[0],tml.index.values[0])
                ,xy=(tml.index.values,tml['KM_estimate'].values), xycoords='data'
                ,xytext=(6,-50), textcoords='offset points', color='#007777', fontsize=smlfs
                ,arrowprops={'facecolor':'#007777', 'width':2})

    ax.set_ylim([0,1])
    ax.set_xlim([0,xmax])
    
def days_between(dy):
    if (dy[0]==dy[0]) and (dy[1]==dy[1]):
        dy[0] = datetime.strptime(dy[0], "%m/%d/%Y")
        dy[1] = datetime.strptime(dy[1], "%m/%d/%Y")
        return (dy[1] - dy[0]).days
    else:
        return 0
```

###Load Data

The `Crunchbase` API wasn't outputting data in the right format, so I downloaded a `CSV` file containing the site's company info. The file needed to be modified significantly even before it could be manipulated in `Python`. I've attached a link to the file [here](https://drive.google.com/file/d/0Byk_gVxZrDKRUEYtMlFxbk9CaWM/view?usp=sharing).

```python
cb_companies = pd.read_csv("crnch.csv",infer_datetime_format=True).convert_objects(convert_numeric=True)

cb_companies['boot_days'] = cb_companies[['founded_at','first_funding_at']].apply(days_between,axis=1)
cb_companies['fund_days'] = cb_companies[['first_funding_at','last_funding_at']].apply(days_between,axis=1)
cb_companies['without_days'] = cb_companies[['last_funding_at','today']].apply(days_between,axis=1)
cb_companies['age'] = cb_companies[['founded_at','today']].apply(days_between,axis=1)
cb_companies['total_days'] = cb_companies['boot_days'] + cb_companies['fund_days'] + cb_companies['without_days'] 


cb_filtered= cb_companies[(cb_companies.boot_days >= 0) & (cb_companies.fund_days >= 0) & (pd.to_datetime(cb_companies.founded_at)<=pd.datetime(2015,9,9)) & (pd.to_datetime(cb_companies.founded_at)>=pd.datetime(1990,1,1))]

print len(cb_companies)
print len(cb_filtered)
```

```python
cb_filtered['day_of_year'] = cb_filtered['founded_at'].apply(lambda x: (pd.to_datetime(x)).dayofyear)
cb_filtered["rankled"] = pd.qcut(cb_filtered.boot_days, 5, labels=["1st","2nd","3rd","4th","5th"])
cb_filtered['failure'] = np.where((cb_filtered['status']=='closed') | np.logical_and(cb_filtered['status']=='operating',cb_filtered['without_days']>=20*30), 1, 0)
cb_filtered.head()
```

Most of the columns should be self-explanatory: 

+ `founded_at` is the date the startup was founded on. Similarly, `first_funding_at` and `last_funding_at` are the dates of the first funding round and most recent funding round, respectively.

+ `boot_days = first_funding_at - founded_at`, or the number of days before the startup raised its first round.

+ `fund_days = last_funding_at - first_funding_at`

+ `without_days = (Sept. 9th, 2015) - last_funding_at`

+ `age = (Sept. 9th, 2015) - first_funding_at`. Note that this should be the same as `total_days`, which is `boot_days + fund_days + without_days`. That said, due to missing or incomplete data, `age` is often different or absent entirely.

+ `day_of_year` is the day of the year on which the company was founded

+ `rankled` is the quintile (1-5) of `boot_days` to which the company belongs.

+ `failure` is 1 if a startup is dead and 0 if it is still alive. [CB Insights](https://www.cbinsights.com/blog/startup-death-data/) found that the average startup dies 20 months after its last funding round (i.e. goes for 20 months unfunded). Accordingly, I defined a dead startup as one whose status is `closed` or still listed as `operating` but $($ `without_days` $/ 30) > 20$.

In terms of filtering, I decided to remove any companies that didn't list their founding date and/or didn't have dates in the order such that `fund_days` and `boot_days` >= 0. To clarify the logic here, `boot_days` can only be zero or negative if a startup is listed as receiving funding before it came into existence and `fund_days` can only be zero or negative if a startup's last round is listed as being raised before its first, both of which are impossible. 

I also removed any companies that were founded before 1990 or after the present day; the majority of companies with `founded_at` entries before 1990 are errors, as are companies with entries in the future.

A frighteningly large number of startup entries (roughly $\frac{1}{3}$, to be exact) violated these logical rules, making the quality of this data highly suspect. If I had to make an educated guess, I would assume that the listings without starting dates most likely belong to younger companies that closed before they ever had any real traction. Without these taking into account these companies, our dataset will likely overstate the survival rate of startups— something to keep in mind as we look at our results.

###Comparing Subsets with Original Dataset Using `logrank_test` 

The process of cleaning a dataset has the potential to introduce unintended biases when one or more of our variables is cross-correlated. For example, earlier entries in the original `Crunchbase` dataset may be less likely to have a defined `founded_at` field. Accordingly, filtering out entries where `founded_at=NaN` unintentionally biases our subset towards later company entries. If our subset is sufficiently biased, then it can no longer serve as an accurate representation of the original dataset (and, by extension, the real world), making any insights drawn from the subset completely worthless. 

The following code use a `logrank_test` method to analyze whether our subsets are sufficiently different from the initial dataset.

```python
cb_total = cb_companies[pd.notnull(cb_companies.founded_at)]
cb_total = cb_total[~cb_total.isin(cb_filtered)].dropna(how='all')
cb_total['total_days'] = cb_total['boot_days'] + cb_total['fund_days'] + cb_total['without_days']
cb_total['failure'] = np.where((cb_total['status']=='closed') | np.logical_and(cb_total['status']=='operating',cb_total['without_days']>=20*30), 1, 0)

results1 = logrank_test(cb_total['age'], cb_filtered['total_days'],cb_total['failure'], cb_filtered['failure'])
results1.summary
```

```python
cb_total = cb_companies[pd.notnull(cb_companies.founded_at)]
cb_total = cb_total[~cb_total.isin(cb_filtered_mark)].dropna(how='all')
cb_total['total_days'] = cb_total['boot_days'] + cb_total['fund_days'] + cb_total['without_days']
cb_total['failure'] = np.where((cb_total['status']=='closed') | np.logical_and(cb_total['status']=='operating',cb_total['without_days']>=20*30), 1, 0)

results2 = logrank_test(cb_total['age'], cb_filtered_mark['total_days'],cb_total['failure'], cb_filtered_mark['failure'])
results2.summary
```

`logrank` measures and reports on whether two intensity processes are different. That is, given two event series, this function  determines whether the data generating processes are statistically different using a chi-squared test statistic. The null hypothesis is that the processes are the same, which in our case would signal that subsetting our data fromt the larger dataset has not changed its overall composition.

The fact that both of these tests reject the null hypothesis with extreme certainty ($p<<0.05$) suggests that our subsets are significantly different from the larger dataset. This would normally put an end to our analysis, seeing as any results we glean are difficult to use in predicting or describing the real world. That said, we have already accepted that `Crunchbase`'s data is a poor reflection of the real world. Given the obvious inaccuracies in the initial dataset, filtering our data may have even made it *more* reflective of the real world. Regardless, these test results shouldn't deter us from continuing with our analyses.

##*b. Cox Proportional Hazards Modeling*

###Transform Raw Dataframe Into Design Matrix

We will use the `patsy` package to transform our row-format dataframe into $X$ using a model-specification. Note that by default, the first factor in each feature will be used in the intercept, which sets the baseline hazard rate, to which all other factor combinations are compared. 

```python
modelspec = 'market'

#I further filtered the dataset to remove market values that showed up less than 35 times. 
#Using the full dataset was too computationally intensive.
cb_filtered_mark_x = cb_filtered_mark[cb_filtered_mark['market'].isin(market_list[market_list > 35].index)]
cb_filtered_mark_t = pt.dmatrix(modelspec, cb_filtered_mark_x, return_type='dataframe')
design_info = cb_filtered_mark_t.design_info
cb_filtered_mark_t = cb_filtered_mark_t.join(cb_filtered_mark_x[['total_days','failure']])

#CoxPHFitter expects reduced-rank design matrix WITHOUT intercept (this took me an embarassingly
# long time to figure out).
del cb_filtered_mark_t['Intercept']
cb_filtered_mark_t.head().T
```

Each column in the matrix above is a company. For each column, the `market[XXXX]` row is either 0 or 1 depending on whether the company belongs to the given market or not, respectively.

###Fit CoxPH Model

```python
cx = sa.CoxPHFitter(normalize=False)
cx.fit(df=cb_filtered_mark_t, duration_col='total_days', event_col='failure'
           ,show_progress=True, include_likelihood=True)
```

The `fit()` method has run the CoxPH partial regression and stored the results on the cx object.

###Baseline Hazard Rate

```python
fig, axes = plt.subplots(nrows=1, ncols=2, squeeze=False, sharex=False)
cx.baseline_cumulative_hazard_.plot(ax=axes[0,0], legend=False
                ,title='Baseline cumulative hazard rate',ylim=[0,1])
cx.baseline_survival_.plot(ax=axes[0,1], legend=False
                ,title='Baseline survival rate',xlim=[0,15000])
```

Here we see our baseline cumulative hazard rate $\Lambda=\sum_i{\lambda_0 (t)}$ and the computed baseline survival rate $S=e^{-\Lambda}$.

+ The baseline cumulative hazard rate is set by `Advertising`. Like normal regression, Cox PH usually sets $exp(\beta X)=0$ to find the baseline (or intercept). That said, since we're dealing with categorical data, we choose one of our `market` values to set the baseline. __Think about it this way:__ if we're trying to compare the battery failure rates of Android, Windows, and iPhone mobile devices on a *relative* (rather than absolute) basis, we need to establish our baseline rate. In this case, we choose the iPhone as our baseline and phrase our results for the other two devices as "Android phone fail at X% the rate of iPhone" and "Windows phones fail at Y% the rate of iPhone."

+ Reading the cumulative baseline hazard rate, we see for example that by ~1 year elapsed, the baseline probability of startup failure is 80%. This seems high, but of course, the baseline will be modified by $exp(\beta X)=0$.

###Summary of Model Coefficients

```python
cx.summary.sort('coef')
```

Once again, we need to keep in mind the proportional nature of the hazard coefficients. For example, `iOS` has an 108% higher hazard rate than `Advertising`, whereas the rate for `Hospitals` is 128% lower.

It's worth noting that the categories with the smallest hazard rates tend to be more traditional industries with higher barriers to entry, whereas those with high rates are much younger with lower barriers to entry.

###$exp(\beta)$ Coefficient Ranges

```python
smy = cx.summary.copy()
smy['err'] = smy['upper 0.95'] - smy['coef']
fig = plt.figure()
axes = fig.gca()
fig.canvas.draw()
axes.errorbar(x=np.arange(smy.shape[0]), y=np.exp(smy['coef'])
                       ,marker='o', linestyle='', yerr=np.exp(smy['err']))
axes.set_title('exp(beta) coefficients')
axes.set_xlim([-0.1, len(smy) - 0.9])
```

Our model shows *HUGE* variance. As such, we should take our results with a grain of salt.

###Evaluate Model Performance Using Concordance

The `lifelines` package has a couple of convenience functions to let us very easily evaluate model performance:

- `concordance_index` which computes the concordance index - the overall probability that predicted and actual events occur in the same ranked order

- `k_fold_cross_validation` which divides the dataset n into two subsets with $n(k-1)/k$ rows and $n/k$ rows, then uses the larger set to train and the smaller set to test the model using the concordance measure. The 'fold' part means it runs repeatedly over different subsets so that each row in the data is used for training and for test.

Paraphrasing the `lifelines docs`, "concordance is a generalisation of the area-under-curve (AUC) measure and interpreted similarly":

- 0.5 is the expected result from random predictions

- 1.0 is perfect concordance

- 0.0 is perfect anti-concordance (multiply predictions with -1 to get 1.0)

A concordance score between 0.6 and 0.8 would indicate a well-performing model.

```python
cx1 = sa.CoxPHFitter(normalize=False)
scores = k_fold_cross_validation(cx1, cb_filtered_mark_t, k=5
        ,duration_col='total_days',event_col='failure', predictor='predict_expectation')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(10,2))
sns.boxplot(scores, orient="h", color='lightblue')
sns.despine(trim=True)
```

Somewhat surprisingly, our model is within the "well-performing" range! While I'm still a little skeptical about the absolute accuracy of our analysis, this is definitely a promising result.

###Transform Raw Dataframe Into Design Matrix

```python
modelspec = 'funding_total_usd + funding_rounds + boot_days'

cb_filtered_t = pt.dmatrix(modelspec, cb_filtered, return_type='dataframe')
design_info = cb_filtered_t.design_info
cb_filtered_t = cb_filtered_t.join(cb_filtered[['total_days','failure']])

## NOTE: CoxPHFitter expects reduced-rank design matrix WITHOUT intercept
## https://courses.nus.edu.sg/course/stacar/internet/st3242/handouts/notes3.pdf
del cb_filtered_t['Intercept']
cb_filtered_t.head().T
```

```python
cx = sa.CoxPHFitter(normalize=False)
cx.fit(df=cb_filtered_t, duration_col='total_days', event_col='failure'
           ,show_progress=True, include_likelihood=True)
```

Instead of referring to a specific category (as in our `market`-based analysis), the coefficients here represent the marginal impact of an additional dollar of funding or another year without funding. This also illustrates a weakness with the CoxPH model mentioned in the __Intro__: the assumption that the marginal impact of an additional dollar/funding round/day before funding stays constant over time and after X number of previous rounds or dollars.

##Summary of Findings
--------------------------------

The quality of our findings is naturally dependent on that of our original dataset. Accordingly, each of our insights is conditioned on a certain set of constraints. To restate our original inquiries: 

1. How do failure rates differ between markets/spaces?

*Younger, smaller markets with lower barriers to entry (e.g. `iOS`, `Social Commerce`, and `Crowdfunding`) tend to have higher failure rates, while companies in larger, more traditional spaces (`Hospitals`, `Food Processing`, and `Medical Devices`) with higher barriers to entry appear to live longer. That said, more traditional industries often scale at a slower rate and are less dependent on consumer marketing. Therefore, it's more likely that they'll enter their information into Crunchbase later on in their lifespan when there's less risk of failure, meaning that survivorship bias factors into our results.*

*Another interesting trend is the relationship between market size and linearity of survival function. Based on our Kaplan-Meier Estimates, it looks as though larger markets don't face the same "drop-off" in failure rates that we notice in more niche spaces.*

2. How does the length of time between a startup's founding and first round affect its probability of failing?

*There seems to be a positive correlation between the number of days a startup waits before raising its first round (as represented by `boot_days` and `rankled`) and survival rate. As previously mentioned however, there is already a natural correlation between `boot_days`, `rankled`, and `total_days`, meaning that our results don't tell us much. What __is__ interesting to observe is how the marginal effect of `boot_days` on `total_days` seems to increase with quintile, suggesting that the relationship between number of days a startup waits and survival rates is non-linear (e.g. the effect of another day without funding on surival rates is greater when a startup has already lasted 1,000 days without funding than when a startup is only at 100 days). Of course, this also assumes that `boot_days` doesn't contain any high outliers that could bias the 5th quintile.*

3. How does funding amount and the number of rounds raised affect failure rates?

*Funding amount and frequency both appear to positively impact the survival rates of startups. That said, our model assumes that the impact of each variable stays constant over time, meaning that we can't account for the fact that funding amount and frequency may play a greater role in startup survival during different stages in the company's lifespan. Still, these results validate the purpose and impact of venture capital investing, which has occasionally been challenged in recent years.*
